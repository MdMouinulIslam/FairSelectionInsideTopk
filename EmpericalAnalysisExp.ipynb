```python
datasetName = "airbnb"
```

```python
numberofSample = 1000
```

```python
dataset = pd.read_csv(r'data/imdb.csv', nrows=numberofSample)
dataset.columns = ['id','name','year','rating','votes','certificate','runtime','genre','metascore','more','director','actor1','actor2',
        'actor3','actor4','collection']
```

```python
lst_col = 'genre' 

dataset = dataset.assign(**{lst_col:dataset[lst_col].str.split(',').apply(lambda x: x[0]) })
```

```python
def queryPredicate(dataset,predicateMap):
    if len(predicateMap) == 0:
        result = dataset.head(1)['id']
        return result
    for predicate, value in predicateMap.items():
        dataset = dataset[dataset[predicate]==value]
    result = (dataset.head(1)['id'],dataset.head(1)['name'])
    return result
```

```python
########################################### no predicate ################################################
predicateMap = []
Test_case = 10
rtime = 0
for i in range(Test_case):
    start = timeit.default_timer()
    QueryPredicate(dataset,predicateMap)
    end = timeit.default_timer()
    rtime = rtime + end - start
print('runtime = ',rtime/Test_case)
```
